```python
tokenizer = Tokenizer()
tokenizer.from_pretrained("./ramayana_hindi.json")
```

```python
# check if encoding and decoding works as expected
tokenizer.decode(tokenizer.encode("हैलो, क्या हाल हैं?"))
```

```python
text_data = open("../data/ramayana.txt", "r").read()
data = torch.tensor(tokenizer.encode(text_data))
len(text_data), len(data)
```

```python
data.shape
```

```python
block_size = 8
```

```python
batch_data = data[:block_size + 1]
xb = batch_data[:block_size]
yb = batch_data[1:]
```

```python
len(xb), len(yb)
```

```python
print(xb)
print(yb)
```

```python
for i in range(block_size):
    print(f"Contenxt is: {xb[:i + 1]}, Target is: {yb[i]}")
```

```python
train_data = data[:int(0.9*len(data))]
val_data = data[int(0.9*len(data)):]
print(f"Len of train data: {len(train_data)}, Len of val data: {len(val_data)}")
```

```python
torch.manual_seed(42)
batch_size = 4
def get_batch(batch_size = batch_size, data_type = "train"):
    data = train_data if data_type == "train" else val_data
    ix = torch.randint(len(data) - block_size, (batch_size,))
    print(ix)
    xb = torch.stack([data[i : i + block_size] for i in ix])
    yb = torch.stack([data[i + 1 : i + 1 + block_size] for i in ix])
    return xb, yb
```

```python
xb, yb = get_batch()
```

```python
print(xb)
print(yb)
```

```python
for row in range(batch_size):
    for i in range(1, block_size):
        print(f"Context is : {xb[row][:i]} Target is : {yb[row][i]}")
```
