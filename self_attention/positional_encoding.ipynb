# Sinosudal encoding

```python
import torch
import matplotlib.pyplot as plt
```

```python
max_dim = 50
dims_ev = torch.arange(0, max_dim, 2)
dims_od = torch.arange(1, max_dim, 2)

d_model = 256
pos = 1
```

```python
vals_ev = torch.sin(pos * dims_ev/d_model).tolist()
vals_od = torch.cos(pos * dims_od/d_model).tolist()
```

```python
len(dims_ev), len(dims_od)
```

```python
vals_final = []
dims_final = []

i = 0
while len(vals_final) < len(dims_ev) + len(dims_od):
    vals_final.append(vals_ev[i])
    vals_final.append(vals_od[i])
    
    dims_final.append(int(dims_ev[i]))
    dims_final.append(int(dims_od[i]))
    i += 1
```

```python
plt.plot(dims_final, vals_final)
plt.show()
```
